```python
# Import necessary libraries
import os
import shutil
from sklearn.model_selection import train_test_split

# Assume dataset is downloaded and extracted
dataset_dir = "path_to_fruits_360_dataset"
train_dir = "fruits_data/train"
test_dir = "fruits_data/test"

# Create directories for organized datasets
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Organize the dataset into train/test folders
classes = os.listdir(dataset_dir)
for fruit_class in classes:
    class_path = os.path.join(dataset_dir, fruit_class)
    images = os.listdir(class_path)
    train, test = train_test_split(images, test_size=0.2, random_state=42)
    
    os.makedirs(os.path.join(train_dir, fruit_class), exist_ok=True)
    os.makedirs(os.path.join(test_dir, fruit_class), exist_ok=True)

    for img in train:
        shutil.copy(os.path.join(class_path, img), os.path.join(train_dir, fruit_class, img))
    for img in test:
        shutil.copy(os.path.join(class_path, img), os.path.join(test_dir, fruit_class, img))


import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set image dimensions
IMG_HEIGHT, IMG_WIDTH = 224, 224
BATCH_SIZE = 32

# Preprocessing with data augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1.0/255)

# Load data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)


from skimage.feature import hog
from skimage.io import imread
from skimage.color import rgb2gray

def extract_hog_features(image_path):
    image = rgb2gray(imread(image_path))
    features, _ = hog(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)
    return features

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Correct the classes variable
classes = list(train_generator.class_indices.keys())

# Adjust the model output layer
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(classes), activation='softmax')  # Adjust the number of classes
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


EPOCHS = 1

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=test_generator
)

import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate on test set
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy}")

# Confusion Matrix
y_true = test_generator.classes
y_pred = model.predict(test_generator)
y_pred_classes = y_pred.argmax(axis=1)

# print(classification_report(y_true, y_pred_classes, target_names=classes))

conf_matrix = confusion_matrix(y_true, y_pred_classes)
plt.imshow(conf_matrix, cmap='Blues')
plt.colorbar()
plt.show()

# Save the model
model.save('fruit_classifier.h5')

# Load and use the model for prediction
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

loaded_model = load_model('fruit_classifier.h5')

def predict_image(image_path):
    img = load_img(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    prediction = loaded_model.predict(img_array)
    predicted_class = classes[np.argmax(prediction)]
    return predicted_class

print(predict_image('images.jpg'))

```
